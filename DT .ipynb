```python
# =====================================================
# STEP 4 — IMPORT MODELS & METRICS
# =====================================================

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

from xgboost import XGBClassifier

# Metrics
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    confusion_matrix,
    classification_report,
    roc_curve
)

print("Models and metrics imported successfully.")
```

```python
# =====================================================
# MODEL EVALUATION FUNCTION
# =====================================================

def evaluate_model(model, X_train, X_test, y_train, y_test, model_name):

    # Train model
    model.fit(X_train, y_train)

    # Predictions
    y_pred = model.predict(X_test)

    # Probabilities
    y_prob = model.predict_proba(X_test)[:, 1]

    # Metrics
    accuracy = accuracy_score(y_test, y_pred)

    precision = precision_score(y_test, y_pred)

    recall = recall_score(y_test, y_pred)

    f1 = f1_score(y_test, y_pred)

    roc_auc = roc_auc_score(y_test, y_prob)

    # Print results
    print("\n" + "="*60)
    print(f"{model_name} RESULTS")
    print("="*60)

    print(f"Accuracy  : {accuracy:.4f}")
    print(f"Precision : {precision:.4f}")
    print(f"Recall    : {recall:.4f}")
    print(f"F1 Score  : {f1:.4f}")
    print(f"ROC-AUC   : {roc_auc:.4f}")

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)

    print("\nConfusion Matrix:")
    print(cm)

    # Classification Report
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))

    # Return metrics
    return {
        'Model': model_name,
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1': f1,
        'ROC_AUC': roc_auc,
        'Probabilities': y_prob
    }
```

```python
# =====================================================
# RISK CATEGORIZATION
# =====================================================

def categorize_risk(prob):

    if prob < 0.30:
        return 'Low Risk'

    elif prob < 0.70:
        return 'Medium Risk'

    else:
        return 'High Risk'

# Apply risk categories
dt_df['Risk_Category'] = dt_df['Failure_Probability'].apply(categorize_risk)

# Count categories
risk_counts = dt_df['Risk_Category'].value_counts()

print(risk_counts)
```

```python
# =====================================================
# STEP 6 — COST PARAMETERS
# =====================================================

FAILURE_REPAIR_COST = 500

PREVENTIVE_MAINTENANCE_COST = 200

DOWNTIME_COST_PER_HOUR = 100

DOWNTIME_HOURS = 5

TOTAL_FAILURE_COST = (
    FAILURE_REPAIR_COST +
    (DOWNTIME_COST_PER_HOUR * DOWNTIME_HOURS)
)

print("Total Failure Cost:", TOTAL_FAILURE_COST)
```

```python
# =====================================================
# STRATEGY 1 — REACTIVE MAINTENANCE
# =====================================================

reactive_failures = dt_df['Actual_Failure'].sum()

reactive_total_cost = (
    reactive_failures *
    TOTAL_FAILURE_COST
)

print("="*60)
print("REACTIVE MAINTENANCE")
print("="*60)

print("Failures:", reactive_failures)

print("Total Cost:", reactive_total_cost)
```

```python
# =====================================================
# STRATEGY 2 — PREVENTIVE MAINTENANCE
# =====================================================

preventive_machines = dt_df[
    dt_df['Risk_Category'] == 'High Risk'
]

preventive_actions = len(preventive_machines)

# Failures still missed
missed_failures = preventive_machines[
    preventive_machines['Actual_Failure'] == 0
]

remaining_failures = dt_df[
    (dt_df['Risk_Category'] != 'High Risk') &
    (dt_df['Actual_Failure'] == 1)
]

preventive_cost = (
    preventive_actions *
    PREVENTIVE_MAINTENANCE_COST
)

remaining_failure_cost = (
    len(remaining_failures) *
    TOTAL_FAILURE_COST
)

preventive_total_cost = (
    preventive_cost +
    remaining_failure_cost
)

print("="*60)
print("PREVENTIVE MAINTENANCE")
print("="*60)

print("Preventive Actions:", preventive_actions)

print("Remaining Failures:", len(remaining_failures))

print("Total Cost:", preventive_total_cost)
```

```python
# =====================================================
# STRATEGY 3 — BUDGET-CONSTRAINED DT
# =====================================================

budget_limit = 25

top_budget_machines = dt_df.sort_values(
    by='Failure_Probability',
    ascending=False
).head(budget_limit)

budget_preventive_actions = len(top_budget_machines)

# Failures prevented
prevented_failures = top_budget_machines[
    top_budget_machines['Actual_Failure'] == 1
]

# Remaining failures
remaining_budget_failures = dt_df.drop(
    top_budget_machines.index
)

remaining_budget_failures = remaining_budget_failures[
    remaining_budget_failures['Actual_Failure'] == 1
]

budget_maintenance_cost = (
    budget_preventive_actions *
    PREVENTIVE_MAINTENANCE_COST
)

budget_failure_cost = (
    len(remaining_budget_failures) *
    TOTAL_FAILURE_COST
)

budget_total_cost = (
    budget_maintenance_cost +
    budget_failure_cost
)

print("="*60)
print("BUDGET-CONSTRAINED DT")
print("="*60)

print("Preventive Actions:", budget_preventive_actions)

print("Prevented Failures:", len(prevented_failures))

print("Remaining Failures:", len(remaining_budget_failures))

print("Total Cost:", budget_total_cost)
```

```python
# =====================================================
# STRATEGY 4 — DYNAMIC THRESHOLD DT
# =====================================================

dynamic_threshold = 0.50

dynamic_machines = dt_df[
    dt_df['Failure_Probability'] > dynamic_threshold
]

dynamic_actions = len(dynamic_machines)

dynamic_remaining_failures = dt_df[
    (dt_df['Failure_Probability'] <= dynamic_threshold) &
    (dt_df['Actual_Failure'] == 1)
]

dynamic_maintenance_cost = (
    dynamic_actions *
    PREVENTIVE_MAINTENANCE_COST
)

dynamic_failure_cost = (
    len(dynamic_remaining_failures) *
    TOTAL_FAILURE_COST
)

dynamic_total_cost = (
    dynamic_maintenance_cost +
    dynamic_failure_cost
)

print("="*60)
print("DYNAMIC THRESHOLD DT")
print("="*60)

print("Preventive Actions:", dynamic_actions)

print("Remaining Failures:", len(dynamic_remaining_failures))

print("Total Cost:", dynamic_total_cost)
```

```python
# =====================================================
# STRATEGY COMPARISON TABLE
# =====================================================

strategy_results = pd.DataFrame({

    'Strategy': [
        'Reactive',
        'Preventive',
        'Budget-Constrained DT',
        'Dynamic Threshold DT'
    ],

    'Preventive Actions': [
        0,
        preventive_actions,
        budget_preventive_actions,
        dynamic_actions
    ],

    'Remaining Failures': [
        reactive_failures,
        len(remaining_failures),
        len(remaining_budget_failures),
        len(dynamic_remaining_failures)
    ],

    'Total Cost': [
        reactive_total_cost,
        preventive_total_cost,
        budget_total_cost,
        dynamic_total_cost
    ]
})

strategy_results
```

```python
# =====================================================
# THRESHOLD SENSITIVITY ANALYSIS
# =====================================================

thresholds = [0.3, 0.5, 0.7, 0.9]

sensitivity_results = []

for threshold in thresholds:

    threshold_machines = dt_df[
        dt_df['Failure_Probability'] > threshold
    ]

    preventive_actions = len(threshold_machines)

    remaining_failures = dt_df[
        (dt_df['Failure_Probability'] <= threshold) &
        (dt_df['Actual_Failure'] == 1)
    ]

    maintenance_cost = (
        preventive_actions *
        PREVENTIVE_MAINTENANCE_COST
    )

    failure_cost = (
        len(remaining_failures) *
        TOTAL_FAILURE_COST
    )

    total_cost = (
        maintenance_cost +
        failure_cost
    )

    sensitivity_results.append({

        'Threshold': threshold,

        'Preventive Actions': preventive_actions,

        'Remaining Failures': len(remaining_failures),

        'Total Cost': total_cost
    })

# Convert to dataframe
sensitivity_df = pd.DataFrame(sensitivity_results)

print(sensitivity_df)
```

```python
# =====================================================
# THRESHOLD VS COST
# =====================================================

plt.figure(figsize=(8,5))

sns.lineplot(
    x='Threshold',
    y='Total Cost',
    data=sensitivity_df,
    marker='o'
)

plt.title("Threshold Sensitivity vs Operational Cost")

plt.ylabel("Total Cost")

plt.show()
```

```python
# =====================================================
# THRESHOLD VS FAILURES
# =====================================================

plt.figure(figsize=(8,5))

sns.lineplot(
    x='Threshold',
    y='Remaining Failures',
    data=sensitivity_df,
    marker='o'
)

plt.title("Threshold Sensitivity vs Remaining Failures")

plt.ylabel("Remaining Failures")

plt.show()
```
